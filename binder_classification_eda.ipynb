
# Binder Classification — EDA & Imbalanced Learning (IC50 Dataset)

**Goal:** Predict **binders** (IC50 ≤ 2 nM) vs **non‑binders** (> 2 nM) **without leaking labels** (i.e., excluding IC50/pChEMBL as features).  
This notebook demonstrates:
- The **severity of class imbalance**
- The **right metrics** for skewed data (PR‑AUC, F1, MCC, Balanced Accuracy)
- A **balanced Random Forest** baseline using only **physicochemical & efficiency descriptors**


## 1) Setup & Imports

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, precision_recall_curve, roc_curve,
                             auc, matthews_corrcoef, balanced_accuracy_score, ConfusionMatrixDisplay)
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

# Plotting defaults: single-plot figures, no seaborn, no custom colors/styles
plt.rcParams.update({"figure.figsize": (8,5), "axes.grid": True})
```

## 2) Load Dataset

```python

# Update path if needed
DATA_PATH = "ic50.tsv"

df = pd.read_csv(DATA_PATH, sep="\t", low_memory=False)
print("Shape:", df.shape)
df.head()
```


## 3) Create Target (Binders) & Show Class Imbalance

**Definition:**  
- Binder = `IC50 (Standard Value) ≤ 2 nM`  
- Non‑binder = `IC50 > 2 nM`

> **Important:** We will **exclude** `Standard Value` (IC50) and `pChEMBL Value` from features to avoid label leakage.


```python

# Convert key columns
df["Standard Value"] = pd.to_numeric(df.get("Standard Value"), errors="coerce")
df["pChEMBL Value"] = pd.to_numeric(df.get("pChEMBL Value"), errors="coerce")

# Keep rows with an IC50 measurement so the label is defined
df = df.dropna(subset=["Standard Value"])

# Binary label
df["Binder"] = (df["Standard Value"] <= 2).astype(int)

# Class counts & imbalance ratio
class_counts = df["Binder"].value_counts().sort_index()
minority_ratio = class_counts.get(1, 0) / class_counts.sum()

print("Class counts (0=Non-binder, 1=Binder):\n", class_counts.to_string())
print(f"Minority class ratio (binders): {minority_ratio:.4f} (~1 in {int(1/max(minority_ratio,1e-9))} samples)")

# Bar plot: class distribution
plt.figure()
plt.bar(["Non-binder (0)", "Binder (1)"], [class_counts.get(0,0), class_counts.get(1,0)])
plt.title("Class Distribution — Severe Imbalance")
plt.ylabel("Count")
plt.xlabel("Class")
plt.show()
```


## 4) Feature Set (No Leakage) & Pre‑processing

We **exclude** `Standard Value` and `pChEMBL Value` from features.  
We keep **physicochemical** & **efficiency** descriptors that are commonly available before biological testing.


```python

candidate_features = [
    "Molecular Weight", "AlogP", "#RO5 Violations",
    "Ligand Efficiency BEI", "Ligand Efficiency LE",
    "Ligand Efficiency LLE", "Ligand Efficiency SEI"
]

# Convert numerics & keep those with sufficient non-missing coverage
for col in candidate_features:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

features = [c for c in candidate_features if c in df.columns and df[c].notna().mean() > 0.4]
print("Using features:", features)

# Impute numeric with mean (simple baseline)
X = df[features].copy()
X = X.fillna(X.mean())

y = df["Binder"].copy()

# Scale numeric features (benefits distance-based/linear models; neutral for trees)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Stratified split to preserve imbalance ratio
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.25, random_state=42, stratify=y
)

X_train.shape, X_test.shape
```


## 5) Model — Balanced Random Forest

We use `class_weight='balanced'` so the classifier gives **more weight to the minority class** during training without oversampling.


```python

rf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced",
    n_jobs=-1
)
rf.fit(X_train, y_train)

# Predictions and probabilities
y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:, 1]
```


## 6) Evaluation — Metrics for Imbalanced Data

**Why not accuracy?** With ~2–3% binders, a naive model that predicts “non‑binder” for everything can be ~97–98% accurate.

**Better metrics:**
- **Precision** (of predicted binders, how many are true binders?) — guards against false positives (expensive follow‑ups).
- **Recall** (of true binders, how many did we catch?) — guards against false negatives (missed actives).
- **F1‑score** — harmonic mean of precision & recall; robust under skew.
- **PR‑AUC** — area under **Precision–Recall** curve; **gold standard** when positives are rare.
- **Balanced Accuracy** — average of TPR and TNR; compensates for skew.
- **MCC** — Matthews Correlation Coefficient (−1..1); robust single‑number summary for imbalanced sets.


```python

from sklearn.metrics import classification_report

# Core metrics
precision, recall, _ = precision_recall_curve(y_test, y_prob)
pr_auc = auc(recall, precision)
mcc = matthews_corrcoef(y_test, y_pred)
bal_acc = balanced_accuracy_score(y_test, y_pred)

print(classification_report(y_test, y_pred, target_names=["Non-binder", "Binder"]))
print(f"PR-AUC: {pr_auc:.3f}")
print(f"Balanced Accuracy: {bal_acc:.3f}")
print(f"MCC: {mcc:.3f}")

# Optional ROC-AUC (can look optimistic with heavy skew)
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
print(f"ROC-AUC (secondary): {roc_auc:.3f}")
```

### Precision–Recall Curve

```python

plt.figure()
plt.plot(recall, precision, linewidth=2, label=f"PR-AUC = {pr_auc:.3f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall Curve (Binder Classification)")
plt.legend()
plt.show()
```

### Confusion Matrix

```python

disp = ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title("Confusion Matrix — Balanced Random Forest")
plt.show()
```

## 7) Feature Importance (No Leakage)

```python

importances = rf.feature_importances_
order = np.argsort(importances)[::-1]

plt.figure()
plt.barh(np.array(features)[order], importances[order])
plt.gca().invert_yaxis()
plt.xlabel("Relative Importance")
plt.title("Feature Importance — Physicochemical & Efficiency Descriptors")
plt.tight_layout()
plt.show()

pd.DataFrame({"Feature": np.array(features)[order], "Importance": importances[order]})
```


## 8) Takeaways

- The dataset is **severely imbalanced** (binders ≈ 2–3%).  
- We evaluated with **PR‑AUC, F1, MCC, Balanced Accuracy** — metrics that remain informative under skew.  
- A **class‑weighted Random Forest** achieves **high precision with strong recall**, using **non‑leaky features** only.  
- **Most predictive** properties include **LLE**, **Molecular Weight**, and **AlogP**, suggesting **efficient, drug‑like** molecules are more likely to be binders.
